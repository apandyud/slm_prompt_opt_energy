```python
def calculate_calibrated_energy(model_name, params_billions, is_lora, epochs, is_moe, gpu_type="A100"):
    # --- ANCHOR POINT (Phi-3-mini 3.8B, 10 epoch, Dense = 16 GPU óra) ---
    phi_params = 3.8
    phi_epochs = 10
    phi_hours = 16.0  # A100-ason
    
    # --- HARDVER ADATOK ---
    # TDP és PUE a fogyasztáshoz
    gpu_specs = {
        "A100": {"tdp": 400, "perf_ratio": 1.0}, # Bázis
        "H200": {"tdp": 700, "perf_ratio": 3.1}  # A H200 kb. 3x gyorsabb az A100-nál LLM tanításban
    }
    PUE = 1.1
    specs = gpu_specs[gpu_type]

    # --- ARÁNYOSÍTÁS ---
    # Alap arány a paraméterszám alapján
    param_ratio = params_billions / phi_params
    
    # Epoch arány (a 16 óra a 10 epoch-ra vonatkozott)
    epoch_ratio = epochs / phi_epochs
    
    # Módszertani szorzó (LoRA vs Full)
    # Ha a 16 óra LoRA volt, akkor a Full-hoz szorozni kell (kb 1.6x)
    # Feltételezzük, hogy a 16 óra a LoRA volt a Phi-nél
    method_multiplier = 1.0 if is_lora else 1.6
    
    # Architektúra szorzó (MoE esetén csak ~25% az aktív számítás)
    arch_multiplier = 0.25 if is_moe else 1.0
    
    # --- IDŐ KISZÁMÍTÁSA (A100 órában) ---
    estimated_hours_a100 = phi_hours * param_ratio * epoch_ratio * method_multiplier * arch_multiplier
    
    # Átszámítás a cél GPU sebességére
    final_hours = estimated_hours_a100 / specs["perf_ratio"]
    
    # --- ENERGIA ---
    energy_wh = final_hours * (specs["tdp"] * PUE)
    
    return {
        "model": model_name,
        "hours": final_hours,
        "energy_wh": energy_wh
    }
```

```python
# --- SZCENÁRIÓK ---
scenarios = [
    # Név, Param, LoRA, Epoch, MoE
    ("Gemma3N", 3, False, 3, True),
    ("Gemma3N", 3, True, 10, True),
    ("Qwen 3.5 4B", 4, False, 3, True),
    ("Qwen 3.5 4B", 4, True, 10, True),
    ("Qwen 3.5 9B", 9, False, 3, True),
    ("Qwen 3.5 9B", 9, True, 10, True),
    ("Phi-3-mini", 3.8, True, 10, False), # Ez az anchor
    ("Llama2 13B", 13, False, 3, False),
    ("Llama2 70B", 70, False, 3, False)
]

print(f"{'Modell':<20} | {'Mód':<6} | {'H':<6} | {'Energia (Wh)':<10}")
print("-" * 55)

for name, p, lora, ep, moe in scenarios:
    res = calculate_calibrated_energy(name, p, lora, ep, moe, gpu_type="A100")
    print(f"{res['model']:<20} | {'LoRA' if lora else 'Full':<6} | {res['hours']:<6.2f} | {res['energy_wh']:<10.2f}")
```
