```python
from langchain_community.chat_models import ChatOllama
llm = ChatOllama(model="gemma3n:e4b", temperature = 0.0, top_p = 1, repeat_penalty=1, presence_penalty=0, frequency_penalty=0, num_predict=400)  # vagy bármely más elérhető Ollama modell
```

# Iterations

```python
runs=[('V18', None, None, None),
       ('V19a', 'res/gemma3n_v18.csv', 'res/g3n_errors_18_cv1_2_1.csv', 37),
      ('V20', 'res/gemma3n_v19a_r1.csv', 'res/g3n_errors_19a_cv1_2_1.csv', 50),
      ('V21',  'res/gemma3n_v19a_r1.csv', 'res/g3n_errors_19a_cv1_2_1.csv', 43 ),
      ('V22b', 'res/gemma3n_v21_r1.csv', 'res/g3n_errors_21_cv1_2_1.csv', 22),
      ('V23', 'res/gemma3n_v21_r1.csv', 'res/g3n_errors_21_cv1_2_1.csv', 54)      
     ]

res={}
```

```python

for i in range(10):
    print('Iteration: ', i)
    if i not in res:
        res[i] = {}
    for (prompt_version, results_file, errors_file, cluster_id) in runs:
        if prompt_version not in res[i]:
            messages = prompt_versions_g3n.prompt_versions[prompt_version]
            messages = [(a, t.replace('/no_think','')) for (a,t)  in messages]
            messages[-1] = ( messages[-1][0],  messages[-1][1] + "/no_think" )
            if (results_file and errors_file and cluster_id):
                print(prompt_version, "cluster Error:")
                start_cluster=datetime.datetime.now()
                cga_utils.calc_fix_mcnemar_performance(llm, messages, results_file, errors_file, cluster_id)
                end_cluster=datetime.datetime.now()
                print(start_cluster, end_cluster)
            else:
                start_cluster=None
                end_cluster= None
                
            print(prompt_version, "global Error:")
            start_global=datetime.datetime.now()
            cga_utils.execute_dataset_predictions(llm, messages, trace_messages=False)
            end_global=datetime.datetime.now()
            res[i][prompt_version] = ({"prompt_version": prompt_version, "start_cluster":start_cluster, "end_cluster": end_cluster, "start_global":start_global, "end_global":end_global  })
            print(start_global, end_global)

```

```python
res_e = []

for i in range(10):
    print('Iteration: ', i)
    for row in res[i].values():
        if row['start_cluster'] and row['end_cluster']:
            start = pd.Timestamp(row['start_cluster']).tz_localize("Europe/Budapest")
            end = pd.Timestamp(row['end_cluster']).tz_localize("Europe/Budapest")
            (wh_counter, wh_gauge_sum,wh_trap, delta ) = energy.fetch_gpu_wh(start, end,  "2s")
            res_e.append({"iteration": i, "prompt_version": row["prompt_version"], "start": start, "end": end, "type": "cluster", "wh_counter": wh_counter, "wh_trap": wh_trap, "delta": delta,  "delta_us": delta, "delta_min": delta/(1000000*60), "avg_wh_counter": (wh_counter/delta)*1000000*3600,"avg_wh_trap": (wh_trap/delta)*1000000*3600 })
        if row['start_global'] and row['end_global']:
            start = pd.Timestamp(row['start_global']).tz_localize("Europe/Budapest")
            end = pd.Timestamp(row['end_global']).tz_localize("Europe/Budapest")
            (wh_counter, wh_gauge_sum,wh_trap, delta ) = energy.fetch_gpu_wh(start, end,  "2s")
            res_e.append({ "iteration": i, "prompt_version": row["prompt_version"], "start": start, "end": end, "type": "global", "wh_counter": wh_counter, "wh_trap": wh_trap,  "delta": delta, "delta_us": delta, "delta_min": delta/(1000000*60), "avg_wh_counter": (wh_counter/delta)*1000000*3600 ,"avg_wh_trap": (wh_trap/delta)*1000000*3600 }  )
pd.DataFrame(res_e)
```

```python
pd.DataFrame(res_e).to_csv('res/G3N_energy_consumtion_devset.csv')
```
